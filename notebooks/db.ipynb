```python
import chromadb

client = chromadb.PersistentClient(path="chroma_db")

# List all collections
print(client.list_collections())

# Open your collection and check count
col = client.get_collection("tax_documents")
print(f"Total chunks: {col.count()}")

```

```python
# Peek at the first few entries
peek = col.peek(limit=15)
for doc_id, text, meta in zip(peek["ids"], peek["documents"], peek["metadatas"]):
  print(f"\n[{doc_id}]  source={meta.get('source')}")
  print(f"  {text[:120]}…")
```
